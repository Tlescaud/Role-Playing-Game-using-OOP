```python
class Character:
    def __init__(self, name, hp, attack):
        self.name = name
        self.hp = hp
        self.attack = attack

    def is_alive(self):
        return self.hp > 0

    def take_damage(self, damage):
        self.hp -= damage
        print(f"{self.name} takes {damage} damage. HP left: {self.hp}")

class Enemy(Character):
    pass

class Player(Character):
    def attack_enemy(self, enemy):
        print(f"You attack {enemy.name}!")
        enemy.take_damage(self.attack)

class Game:
    def __init__(self):
        self.player = Player("Hero", 30, 10)
        self.enemy = Enemy("Goblin", 20, 5)

    def play(self):
        print("Welcome to the Dungeon!")
        while self.player.is_alive() and self.enemy.is_alive():
            action = input("Do you want to (A)ttack or (R)un? ").lower()
            if action == 'a':
                self.player.attack_enemy(self.enemy)
                if self.enemy.is_alive():
                    print(f"{self.enemy.name} strikes back!")
                    self.player.take_damage(self.enemy.attack)
            elif action == 'r':
                print("You run away!")
                break
            else:
                print("Invalid action.")

        if not self.player.is_alive():
            print("You have been defeated...")
        elif not self.enemy.is_alive():
            print("Victory! The enemy is vanquished.")

# Start the game
Game().play()
```
